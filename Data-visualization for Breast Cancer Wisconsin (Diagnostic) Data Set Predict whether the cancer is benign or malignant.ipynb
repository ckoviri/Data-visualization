If we see the data, the columns "Unnamed: 32","id" are useless
and column diagnosis is categorical variable and it is our target variable 

So let us remove the columns "Unnamed: 32","id" 
and also seperate target variable and features

```python
# y includes our target and x includes our features
y = data["diagnosis"] 
list = ['Unnamed: 32','id','diagnosis']
x = data.drop(list,axis = 1 )
x.head()
```

```python
sns.violinplot(x["radius_mean"])
```

```python
plt.figure(figsize=(30,15))
sns.boxplot(x="variable", y="value", data=pd.melt(x))
plt.xticks(fontsize=20,rotation=90)
plt.yticks(fontsize=20)
```

```python
sns.distplot(data["area_worst"])
```

#### Why to we need standardization?

```python
data_dia = y
data = x
data_n_2 = (data - data.mean()) / (data.std()) 
```

```python
plt.figure(figsize=(30,15))
sns.boxplot(x="variable", y="value", data=pd.melt(data_n_2))
plt.xticks(fontsize=20,rotation=90)
plt.yticks(fontsize=20)
```

```python
sns.distplot(data_n_2["area_worst"])
```

```python
             # standardization
data = pd.concat([y,data_n_2.iloc[:,:]],axis=1)
data = pd.melt(data,id_vars="diagnosis",var_name="features", value_name='value')
plt.figure(figsize=(30,10))
plt.rc('font', size=20)
sns.violinplot(x="features", y="value", hue="diagnosis", data=data,split=True, inner="quart")
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(30,10))
plt.rc('font', size=20)
sns.boxplot(x="features", y="value", hue="diagnosis", data=data)
plt.xticks(rotation=90)
```
